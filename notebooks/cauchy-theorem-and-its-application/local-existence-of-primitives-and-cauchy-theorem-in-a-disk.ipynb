```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.axes import Axes

from explot import (
    plot_curve,
    plot_directed_line_segment,
    DirectedLineSegmentArrowPosition,
)
```

```python
figures_dir = Path("../../tex/figures")
```

```python
plt.style.use("diagram")

axs: np.ndarray[Axes]
fig, axs = plt.subplots(
    2,
    3,
    sharex=True,
    squeeze=False,
    figsize=(14, 14),
)


z = 1.0 + 1j * 2.0
w = 4.0 + 1j * 6.0

# Row: 0, Column: 0

plot_directed_line_segment(
    0.0,
    w.real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[0, 0],
)

plot_directed_line_segment(
    w.real,
    w,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[0, 0],
)


plot_directed_line_segment(
    z,
    z.real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[0, 0],
)

plot_directed_line_segment(
    z.real,
    0.0,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[0, 0],
)

axs[0, 0].annotate(
    "$0$",
    xy=(0, 0.1),
    fontsize=20,
)

axs[0, 0].annotate(
    "$z$",
    xy=(z.real - 0.1, z.imag + 0.1),
    fontsize=20,
)

axs[0, 0].annotate(
    "$z + h$",
    xy=(w.real - 0.7, w.imag + 0.1),
    fontsize=20,
)

# Row: 0, Column: 1

plot_directed_line_segment(
    z,
    z.real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[0, 1],
)

plot_directed_line_segment(
    z.real,
    w.real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[0, 1],
)

plot_directed_line_segment(
    w.real,
    w,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[0, 1],
)

axs[0, 1].annotate(
    "$z$",
    xy=(z.real - 0.1, z.imag + 0.1),
    fontsize=20,
)

axs[0, 1].annotate(
    "$z + h$",
    xy=(w.real - 0.7, w.imag + 0.1),
    fontsize=20,
)

# Row: 0, Column: 2

plot_directed_line_segment(
    z,
    z.real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[0, 2],
)

plot_directed_line_segment(
    z.real,
    w.real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[0, 2],
)

plot_directed_line_segment(
    w.real,
    w,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[0, 2],
)

plot_directed_line_segment(
    z,
    w.real + 1j * z.imag,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[0, 2],
)

plot_directed_line_segment(
    w.real + 1j * z.imag,
    w.real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[0, 2],
)

plot_directed_line_segment(
    w.real,
    z.real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[0, 2],
)

plot_directed_line_segment(
    z.real,
    z,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[0, 2],
)

axs[0, 2].annotate(
    "$z$",
    xy=(z.real - 0.1, z.imag + 0.1),
    fontsize=20,
)

axs[0, 2].annotate(
    "$z + h$",
    xy=(w.real - 0.7, w.imag + 0.1),
    fontsize=20,
)

# Row: 1, Column: 0

plot_directed_line_segment(
    z,
    z + (w - z).real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[1, 0],
)

plot_directed_line_segment(
    z + (w - z).real,
    w,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[1, 0],
)

axs[1, 0].annotate(
    "$z$",
    xy=(z.real - 0.1, z.imag + 0.1),
    fontsize=20,
)

axs[1, 0].annotate(
    "$z + h$",
    xy=(w.real - 0.7, w.imag + 0.1),
    fontsize=20,
)

# Row: 1, Column: 1

plot_directed_line_segment(
    z,
    z + (w - z).real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[1, 1],
)

plot_directed_line_segment(
    z + (w - z).real,
    w,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    ax=axs[1, 1],
)

plot_directed_line_segment(
    z,
    w,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[1, 1],
)


plot_directed_line_segment(
    w,
    z + (w - z).real,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[1, 1],
)

plot_directed_line_segment(
    z + (w - z).real,
    z,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.RIGHT,
    color="orange",
    linestyle="--",
    ax=axs[1, 1],
)

axs[1, 1].annotate(
    "$z$",
    xy=(z.real - 0.1, z.imag + 0.1),
    fontsize=20,
)

axs[1, 1].annotate(
    "$z + h$",
    xy=(w.real - 0.7, w.imag + 0.1),
    fontsize=20,
)

# Row: 1, Column: 2

plot_directed_line_segment(
    z,
    w,
    arrow_distance=0.2,
    arrow_length=0.5,
    arrow_position=DirectedLineSegmentArrowPosition.LEFT,
    ax=axs[1, 2],
)

axs[1, 2].annotate(
    "$z$",
    xy=(z.real - 0.1, z.imag + 0.1),
    fontsize=20,
)

axs[1, 2].annotate(
    "$z + h$",
    xy=(w.real - 0.7, w.imag + 0.1),
    fontsize=20,
)

axs[1, 2].annotate(
    "$\\eta$",
    xy=(((z + w) / 2).real + 0.2, ((z + w) / 2).imag - 0.2),
    fontsize=20,
)

ax: Axes
for ax in axs.flatten():
    ax.axis("equal")


fig.savefig(
    figures_dir.joinpath("adding-a-rectangle-and-a-triangle.png"),
    bbox_inches="tight",
)
plt.show()
```
